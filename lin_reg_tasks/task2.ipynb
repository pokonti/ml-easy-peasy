```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df.year, df['per capita income (US$)'], color='green');
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.coef_ 
```

```python
reg.intercept_
```

```python
prediction = reg.predict(pd.DataFrame({'year': [2020]}))
print(prediction)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df.year, df['per capita income (US$)'], color='green')
plt.plot(df.year, reg.predict(df[['year']]), color='black');
```
